```python
import json
with open('BTC_price.json', 'r') as json_file:
    dataset = json.load(json_file)
# Extract the relevant information from the dataset
coin_data = dataset['coins']['coingecko:bitcoin']
timestamps = [entry['timestamp'] for entry in coin_data['prices']]
prices = [entry['price'] for entry in coin_data['prices']]

# Create a DataFrame
df = pd.DataFrame({'timestamp': timestamps, 'price': prices})

# Convert the timestamp to a datetime object
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s').dt.date
```

```python
df
```

```python
import pandas as pd

# Convert timestamp in btc_df to datetime
btc_df['timestamp'] = pd.to_datetime(btc_df['timestamp'])

# Convert timestamp in df to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Merge the two DataFrames on the 'timestamp' column
merged_df = pd.merge(btc_df, df, on='timestamp', how='inner')

# Display the merged DataFrame
print(merged_df)
```

```python
# Calculate the 50-day moving average
merged_df['moving_average_50d'] =merged_df['price'].rolling(window=50).mean()

# Calculate the daily percentage change of the moving average
merged_df['momentum_50d'] = merged_df['moving_average_50d'].pct_change() * 100
```

```python
merged_df
```

```python
merged_df
```

```python
#Applying 30% weight to dominance
merged_df['weighted_dominance'] = merged_df['dominance'] * 0.3
```

```python
merged_df
```

```python
#Calculating the chagne in momentum percentage 
merged_df['momentum_change'] = merged_df['momentum_50d'].diff()

#Applying 30% weightage
merged_df['weighted_momentum_change'] = merged_df['momentum_change'] * 0.3
```

```python
merged_df
```

```python
#Combining the weighed result for final score
merged_df['combined_score'] = merged_df['weighted_dominance'] + merged_df['weighted_momentum_change']
```

```python
merged_df
```

```python
#Function to return the score for a given date
import pandas as pd

# Assuming 'merged_df' is your DataFrame

def get_combined_score(date, df):

    # Filter the DataFrame for the specified date
    selected_row = df[df['timestamp'] == date]

    # Check if the date exists in the DataFrame
    if selected_row.empty:
        return None  

    # Extract the combined score from the selected row
    combined_score = selected_row['combined_score'].values[0]
    
    return combined_score
```

```python
input_date = '2023-12-02'
result = get_combined_score(input_date, merged_df)

if result is not None:
    print(f"The combined score for {input_date} is: {result}")
else:
    print(f"No data available for {input_date}")
```
